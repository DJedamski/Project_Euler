```python
# 1. Multiples of 3 and 5

list=[]

for i in range(1000):
    if (i % 3 == 0) or (i % 5 == 0):
        list.append(i)
        
sum(list)
```

```python
# 2. Even Fibonacci Numbers

Seq=[1,2]

for n in Seq:
    if n > 1:
        if Seq[-1] <= 4000000:
            Seq.append(Seq[-1] + Seq[-2])

evens=[]

for n in Seq:
    if n % 2 == 0:
        evens.append(n)      

print Seq
print evens

sum(evens) 
```

```python
# 3. Largest Prime Factor

def Factor(num):
    largestFactor = 0
    factor = 2
    product = 1

    while product != num:
        if num % factor == 0:
            largestFactor = factor
            product *= factor
        factor += 1

    return largestFactor


print Factor(600851475143)
```
